```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
quantization_config = BitsAndBytesConfig(load_in_4bit=True)
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    quantization_config=quantization_config
)
```

```python
def generate_text(prompt, tokenizer, model, max_new_tokens=256):
    tokenizer.pad_token = tokenizer.eos_token
    inputs = tokenizer(prompt, return_tensors="pt", padding=True, truncation=True)
    inputs = {key: value.to(model.device) for key, value in inputs.items()}

    output = model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        do_sample=True,
        repetition_penalty=1.2
    )

    # ✅ **只提取 Assistant 部分**
    response = tokenizer.decode(output[0], skip_special_tokens=True)
    response = response.split("Assistant:")[-1].strip()  # 确保不返回 Context
    return response
```

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

embedding_model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
documents = [
    "AI is used in radiology for medical image analysis, including X-rays, CT scans, and MRIs.",
    "Deep-learning models such as CNN and U-Net are widely used in radiology for tumor detection and segmentation.",
    "放射学中的人工智能常用方法包括卷积神经网络（CNN）、U-Net 分割模型、迁移学习等。",
    "基于 Transformer 的医学影像分析正在兴起，能够提供更强的上下文理解能力。",
    "常见的 AI 影像分析工具包括自动病灶检测、图像分割、计算机辅助诊断（CAD）等。",
    "AI 在放射学中的挑战：数据隐私、模型可解释性、对医学数据的高需求。",
]


doc_embeddings = embedding_model.encode(documents)
index = faiss.IndexFlatL2(doc_embeddings.shape[1])
index.add(np.array(doc_embeddings, dtype=np.float16))
```

```python
import re
def detect_language(text):
    """ 简单判断是中文还是英文 """
    if re.search("[\u4e00-\u9fff]", text):  # 只要有一个中文字符，就认为是中文
        return "chinese"
    return "english"

def retrieve_relevant_docs(query, top_k=2):
    lang = detect_language(query)  # 识别问题的语言
    context_results = retrieve_relevant_docs(query, top_k=top_k)

    # 过滤 context，确保与 query 语言匹配
    filtered_docs = [
        doc for doc in context_results
        if (lang == "chinese" and any("\u4e00" <= char <= "\u9fff" for char in doc)) or
           (lang == "english" and all("\u4e00" > char or char > "\u9fff" for char in doc))
    ]

    # 如果有匹配的语言文档，就返回，否则返回默认检索结果
    return filtered_docs if filtered_docs else context_results

def retrieve_relevant_docs(query, top_k=2):
    query_embedding = embedding_model.encode([query])
    distances, indices = index.search(np.array(query_embedding, dtype=np.float32), top_k)
    return [documents[i] for i in indices[0]]

def rag_chatbot(query):
    lang = detect_language(query)
    context = retrieve_relevant_docs(query)
    context_text = "\n".join(context)

    if lang == "english":
        prompt = f"Context: {context_text}\nBased on the above information, please provide a concise and professional answer in English. Avoid repeating the original text.\nUser: {query}\nAssistant:"
    elif lang == "chinese":
        prompt = f"以下是关于{query}的相关信息：\n{context_text}\n请基于以上信息，用专业、简洁的中文回答，不要重复原文，不要使用英文。\n用户: {query}\n助手:"



    response = generate_text(prompt, tokenizer, model)

    return f"\n📌 **Chatbot Interaction** 📌\n" \
           f"🔹 **Prompt:**\n{prompt}\n" \
           f"🔹 **Context:**\n{context_text}\n" \
           f"🔹 **Question:** {query}\n" \
           f"🔹 **Answer:** {response}\n"
```
